```python
# importing packages
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# importing the heart disease data
heart_df = pd.read_csv('heart-disease.csv')
heart_df.head()
```

```python
heart_df.info()
```

```python
heart_df.describe()
```

```python
# checking the different values in every columns 
columns = list(heart_df.columns)
for i in columns :
    print(heart_df[str(i)].value_counts())

```

```python
# create x (feature matrice)
x = heart_df.drop(columns='target', axis=1)
# create the target (labels)
y = heart_df['target']
```

```python
# choose the right model and hyper parameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
# we'll keep the default hyper parameters 
clf.get_params()
```

```python
# fit the model to the training data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
clf.fit(x_train, y_train)
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
# evaluate the model on the train and test data
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# improve our model 
# try different amount of n_estimators
for i in range(10, 110, 10):
    print(f'trying model with {i} estimators ...')
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f'model accuracy of test set : {clf.score(x_test, y_test)*100:.2f}')
```

```python
# so we use the model with n_estimator = 80
clf = RandomForestClassifier(n_estimators=80)
clf.fit(x_train, y_train)
```

```python
# now we want to save the model and load it 
import pickle
pickle.dump(clf, open('random_forest_model_1.pkl', 'wb'))
```

```python
# importing the model we saved 
loaded_model = pickle.load(open('random_forest_model_1.pkl', mode='rb'))
loaded_model.score(x_test, y_test)
```
